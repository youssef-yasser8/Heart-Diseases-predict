```python
df=pd.read_csv(r"C:\Users\NOUR SOFT\Desktop\New folder\framingham_heart_study.csv")
def classify_blood_pressure(data):

    systolic,diastolic=data["sysBP"],data["diaBP"]
    
    if systolic < 90 or diastolic < 60:
        return "Low (Hypotension)"
    elif 90 <= systolic < 120 and 60 <= diastolic < 80:
        return "Normal"
    elif 120 <= systolic < 130 and diastolic < 80:
        return "Elevated"
    elif 130 <= systolic < 140 or 80 <= diastolic < 90:
        return "High (Hypertension Stage 1)"
    elif 140 <= systolic < 180 or 90 <= diastolic < 120:
        return "High (Hypertension Stage 2)"
    else:
        return "Hypertensive Crisis (Seek Medical Attention)"


df["BP_Cat"] = df.apply(classify_blood_pressure, axis=1)
def classifyBMI(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal Weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obese (Class 1)'
    elif 35 <= bmi < 40:
        return 'Obese (Class 2)'
    else:
        return 'Obese (Class 3)'

df["BMI_category"]=df["BMI"].apply(classifyBMI)
def classifyHeartRate(hr):
    if hr < 60:
        return 'Bradycardia (Low)'
    elif 60 <= hr <= 100:
        return 'Normal'
    else:
        return 'Tachycardia (High)'

df["heartRate_cat"]=df["heartRate"].apply(classifyHeartRate)
def classifyGlucose(glucose):
    if glucose < 100:
        return 'Normal'
    elif 100 <= glucose < 126:
        return 'Prediabetes'
    else:
        return 'Diabetes'
df["glucose_cat"]=df["glucose"].apply(classifyGlucose)
def makeCholestralCategories(data):
    if data<200:
        return "Desirable"
    elif 200<=data<=239:
        return "BorderLine"
    else:
        return "High"

df["totChol_cat"]=df["totChol"].apply(makeCholestralCategories)
x=df.drop(["TenYearCHD"] ,axis=1)
y=df["TenYearCHD"]
x.loc[[1497, 1625, 3021, 3106, 3177], 'cigsPerDay']=9
x['cigsPerDay'] = x['cigsPerDay'].astype('Int64')
x.loc[[77, 193, 314, 421, 765, 769, 998, 1122, 1284, 1301, 1566, 1573,1616, 1721, 1857, 1913, 1986, 2002, 2120, 2835, 2943, 3226, 
                   3313, 3373, 3375, 3644, 3737, 4008, 4139, 4162, 4235], 'BPMeds']=0
x.loc[[49, 245, 395, 797, 1002, 1044, 1104, 1177, 1206, 1861, 1983, 1985,2074, 2173, 2181, 2367, 2608, 2645, 2738, 3526, 3791, 3816], 'BPMeds']=1
x['education'] = x['education'].astype('Int64')
x = x.drop(index=[3160,3673,3927,2657,3008,339,  358, 1022, 2237, 2482, 2492, 2663, 2922, 3142, 3225, 3747, 3862, 3901, 4053, 4070 , 2406, 3749, 3849, 3971])
y = y.drop(index=[3160,3673,3927,2657,3008,339,  358, 1022, 2237, 2482, 2492, 2663, 2922, 3142, 3225, 3747, 3862, 3901, 4053, 4070 , 2406, 3749, 3849, 3971])
```

```python
# best model
from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline as imblearn_pipeline
edu_pipe = skl_make_pipeline(SimpleImputer(strategy="most_frequent"))
cig_pipe = skl_make_pipeline(SimpleImputer(strategy="mean"))
log_columns = ['glucose', 'sysBP', 'totChol']
log_impute_log = Pipeline([
    ("imputer", KNNImputer(n_neighbors=3)),
    ("log", FunctionTransformer(np.log1p))
])

impute_only_columns = ['BMI', 'heartRate']
impute_only_pipe = Pipeline([
    ("imputer", KNNImputer(n_neighbors=3))
])

ordinal_features = [
    "BMI_category",
    "heartRate_cat",
    "glucose_cat",
    "totChol_cat",
    "BP_Cat"
]

ordinal_categories = [
    ["Underweight","Normal Weight","Overweight","Obese (Class 1)","Obese (Class 2)","Obese (Class 3)"],
    ["Bradycardia (Low)","Normal","Tachycardia (High)"],
    ['Normal', 'Prediabetes', 'Diabetes'],
    ["Desirable","BorderLine","High"],
    ["Low (Hypotension)",
     "Normal",
     "Elevated",
     "High (Hypertension Stage 1)",
     "High (Hypertension Stage 2)",
     "Hypertensive Crisis (Seek Medical Attention)"]
]

missing_and_preprocessing = ColumnTransformer(transformers=[
    ("edu_pipe", edu_pipe, ["education"]),
    ("cig_pipe", cig_pipe, ["cigsPerDay"]),
    ("log_impute_log", log_impute_log, log_columns),
    ("impute_only", impute_only_pipe, impute_only_columns),
    ("ordinal", OrdinalEncoder(categories=ordinal_categories), ordinal_features)
])

logreg_pipeline = imblearn_pipeline(steps=[
    ('preprocessing', missing_and_preprocessing),
    ('scaler', StandardScaler()),
    ('under', RandomUnderSampler(sampling_strategy={0: 1000}, random_state=42)),
    ('feature_selection', SelectFromModel(
        RandomForestClassifier(n_estimators=100, random_state=42),
        max_features=7,
        threshold=-np.inf
    )),
    ('model', LogisticRegression(solver='liblinear', class_weight='balanced'))
])

param_grid_logreg = {
    'model__C': [0.01, 0.1, 1, 10],
    'model__penalty': ['l1', 'l2']
}

grid_logreg = GridSearchCV(
    estimator=logreg_pipeline,
    param_grid=param_grid_logreg,
    scoring='f1',
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    n_jobs=-1,
    verbose=1
)

grid_logreg.fit(x, y)

print("\nBest Parameters for LogisticRegression:")
print(grid_logreg.best_params_)

best_logreg_model = grid_logreg.best_estimator_

cv_logreg = cross_validate(
    estimator=best_logreg_model,
    X=x,
    y=y,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring=["accuracy", "recall", "precision", "f1"],
    return_train_score=True,
    n_jobs=-1
)

print("\nFinal Evaluation for LogisticRegression (After Tuning):")
print(f"Train Accuracy: {cv_logreg['train_accuracy'].mean():.3f}")
print(f"Test Accuracy:  {cv_logreg['test_accuracy'].mean():.3f}")
print(f"Train Recall:   {cv_logreg['train_recall'].mean():.3f}")
print(f"Test Recall:    {cv_logreg['test_recall'].mean():.3f}")
print(f"Train Precision:{cv_logreg['train_precision'].mean():.3f}")
print(f"Test Precision: {cv_logreg['test_precision'].mean():.3f}")
print(f"Train F1:       {cv_logreg['train_f1'].mean():.3f}")
print(f"Test F1:        {cv_logreg['test_f1'].mean():.3f}")
```
